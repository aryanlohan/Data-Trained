# Training process begins

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
```

```python
maxAccu=0 
maxRS=0
```

```python
for i in range(1,200):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=i)
    LR = LogisticRegression()
    LR.fit(x_train,y_train)   
    predrf=LR.predict(x_test)  
    acc=accuracy_score(y_test,predrf)  
    print("accuracy: ",acc, "random_state: ",i)
    if acc>maxAccu:
        maxAccu=acc
        maxRS=i
        print("accuracy: ",maxAccu, "random_state: ",i)
```

```python
print("Best accuracy is", maxAccu,"on Random State", maxRS)
```

# Creating train_test_split

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20, random_state=173)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

# Logistic Regression

```python
LR = LogisticRegression()
LR.fit(x_train,y_train)
y_pred=LR.predict(x_test)
print("Accuracy", accuracy_score(y_test,y_pred)*100)  #testing accuracy
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_pred_dt=dt.predict(x_test)
print("Accuracy", accuracy_score(y_test,y_pred_dt)*100) #testing accuracy
print(confusion_matrix(y_test,y_pred_dt))
print(classification_report(y_test,y_pred_dt))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred_rf = rf.predict(x_test)
print("Accuracy", accuracy_score(y_test,y_pred_rf)*100) #testing accuracy
print(confusion_matrix(y_test,y_pred_rf))
print(classification_report(y_test,y_pred_rf))
```

# SVC

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(x_train,y_train)

y_pred_svc = svc.predict(x_test)
print("Accuracy", accuracy_score(y_test,y_pred_svc)*100) 
print(confusion_matrix(y_test,y_pred_svc))
print(classification_report(y_test,y_pred_svc))
```

```python
from sklearn.metrics import plot_roc_curve
disp = plot_roc_curve(dt,x_test,y_test)
plot_roc_curve(LR,x_test,y_test,ax=disp.ax_)
plot_roc_curve(svc,x_test,y_test,ax=disp.ax_)
plot_roc_curve(rf,x_test,y_test,ax=disp.ax_)
plt.show()
```
